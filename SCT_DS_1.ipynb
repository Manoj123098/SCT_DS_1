```python
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#  a bar chart with housing_median_age on the x-axis and median_house_value on the y-axis.

import matplotlib.pyplot as plt
plt.title('Value of House Based on its Age')
plt.xlabel('housing_median_age')
plt.ylabel('median_house_value')
plt.bar(df['housing_median_age'], df['median_house_value'])

plt.legend('median_house_value')
plt.show()

```
